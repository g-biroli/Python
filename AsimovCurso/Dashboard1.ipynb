```python
#Importar pandas e os (Operation System)
#Importando ydata_profile para analizar os dados rápido e fácil
import pandas as pd
from ydata_profiling import ProfileReport
import os

#Criação da variavel dos dados
caminho_dados = r"C:\Users\gabri\OneDrive\Belgeler\GitHub\Python\Dados\2025_Viagem.csv"

#Leitura de dados com as opções de máximo de colunas e sem nenhum tipo de restrição
df_viagens = pd.read_csv(caminho_dados, sep=";", encoding="Windows-1252", decimal=",")

#Coluna Despesas criada para junção de gastos totais
df_viagens['Despesas'] = df_viagens['Valor diárias'] + df_viagens['Valor passagens'] + df_viagens['Valor outros gastos']

#Ajustando valores nulos na coluna "Cargos"
df_viagens["Cargo"] = df_viagens["Cargo"].fillna("NÃO IDENTIFICADO")

#Formatação das Datas dos dados
df_viagens['Período - Data de início'] = pd.to_datetime(df_viagens['Período - Data de início'], format="%d/%m/%Y")
df_viagens['Período - Data de fim'] = pd.to_datetime(df_viagens['Período - Data de fim'], format="%d/%m/%Y")

#Criação de tabelas (Nome do Mês da Viagem + Duração da Viagem) 
df_viagens['Mês da viagem'] = df_viagens['Período - Data de início'].dt.month_name()
df_viagens['Periodo da Viagem'] = (df_viagens['Período - Data de fim'] - df_viagens['Período - Data de início'])

```

```python
(df_viagens['Cargo'].value_counts(normalize=True)*100).rename('Proporção de Viagens').reset_index()
```

```python
df_viagens.groupby("Cargo")["Despesas"].max().reset_index().sort_values('Despesas', ascending=False)
```

```python
gastos_totais_por_cargo = df_viagens.groupby("Cargo")["Despesas"].sum().reset_index()
```

```python
df_viagens_consolidadas = (
    df_viagens
    .groupby('Cargo')
    .agg(
        despesa_media = ('Despesas' , 'mean'),
        media_dias_viagem = ('Periodo da Viagem', 'mean'),
        despesas_totais = ('Despesas', 'sum'),
        destino_mais_frequente = ('Destinos', pd.Series.mode),
        n_viagens = ('Nome', 'count')
        )
        .reset_index()
        )
```
